## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
# your code here
from scipy.stats import uniform
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# your code here
def generate_uniform_random_numbers(bottom, ceiling, count):
    random_nums = uniform.rvs(loc=bottom, scale=ceiling-bottom, size=count)
    return random_nums
```

```python
# your code here
#First set of params
plot_1 = generate_uniform_random_numbers(10, 15, 100)

#Second set of params
plot_2 = generate_uniform_random_numbers(10, 60, 1000)
```

```python
# your code here
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot for the first set
axes[0].hist(plot_1, bins=10, edgecolor='black', alpha=0.7)
axes[0].set_title('Uniform Distribution (10 to 15) - 100 Samples')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Count')

# Plot for the second set
axes[1].hist(plot_2, bins=10, edgecolor='black', alpha=0.7)
axes[1].set_title('Uniform Distribution (10 to 60) - 1000 Samples')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Count')

# Display the plot
plt.tight_layout()
plt.show()
```

How are the two distributions different?

### Your answer below
The two uniform distributions differ primarily in their sample size and range. The first distribution, with a smaller sample size of 100 values and a narrow range between 10 and 15, shows more fluctuations in bar heights, resulting in a more "spiky" histogram. In contrast, the second distribution has a larger sample size of 1,000 values and a wider range between 10 and 60. This larger sample size leads to a smoother, more even distribution, with more consistent bar heights. In conclusion, the first distribution appears more variable, while the second distribution is more uniform and less irregular due to the larger sample.

```python
# your code here
def generate_normal_random_numbers(mean, std_dev, count):
    return np.random.normal(loc=mean, scale=std_dev, size=count)
```

```python
# your code here
normal_1 = generate_normal_random_numbers(10, 1, 1000)
```

```python
# your code here
normal_2 = generate_normal_random_numbers(10, 50, 1000)
```

How are the two distributions different?

### Your answer here

The two normal distributions are distinguished by their standard deviation values. The first distribution, with a mean of 10 and a standard deviation of 1, is tightly centered around the mean, resulting in most values being close to 10, creating a narrow and sharply peaked distribution. In contrast, the second distribution, with the same mean of 10 but a standard deviation of 50, shows much greater variability. The values are spread out over a wider range, creating a flatter, broader curve.

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here
def exponential_distributed_numbers(mean, count):
    return np.random.exponential(scale=mean, size=count).tolist()

#Exponential 1
exponential_sequence_1 = exponential_distributed_numbers(10, 10)

#Exponential 2
exponential_sequence_2 = exponential_distributed_numbers(10,100)

#Plot
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot for the first set
axes[0].hist(exponential_sequence_1, bins=100, edgecolor='black', alpha=0.7)
axes[0].set_title('Exponential Distribution Histogram (10 Numbers)')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Count')

# Plot for the second set
axes[1].hist(exponential_sequence_2, bins=100, edgecolor='black', alpha=0.7)
axes[1].set_title('Exponential Distribution Histogram (100 Numbers)')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Count')

# Display the plot
plt.tight_layout()
plt.show()
```

How are the two distributions different?

### Answer
The small sample size (10 numbers) results in a jagged, irregular histogram due to greater random variability, making it less representative of the true exponential distribution. In contrast, the larger sample size (100 numbers) produces a smoother histogram that more accurately reflects the expected exponential decay, demonstrating how larger samples better capture the underlying distribution.

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your answer here
# Hint: This is same as saying P(x<15)

mean_time = 10
lambda_value = 1 / mean_time
time_spent = 15

probability = 1 - math.exp(-lambda_value * time_spent)
probability_percent = probability * 100

print(f"The probability that a customer spends less than 15 minutes in the bank is {probability:.4f} or {round(probability_percent,1)}%")
```
